Ordinal Encoding : Ordinal Encoding is used to convert categorical data with a meaningful order into numerical values. Unlike one-hot encoding, ordinal encoding assigns increasing integers to categories based on their rank.✅ When categories have a natural order (e.g., Low < Medium < High).
🚫 Not suitable for unordered categories (e.g., colors: Red, Green, Blue).
Sklearn and Map Function

```python
# Define size data
data = {
    "Size": ["Small", "Medium", "Large", "Extra Large", "Small", "Medium", "Large", "Extra Large"]
}

# Create DataFrame
size_df = pd.DataFrame(data)
size_df.head(3)
```

```python
# Define order for ordinal encoding
size_order = [["Small", "Medium", "Large", "Extra Large"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=size_order)
oe.fit(size_df[["Size"]])
size_df["Encoding_sklearn"]=oe.transform(size_df[["Size"]])
size_df
```

Using Map Function

```python
# Define manual encoding using map()
size_mapping = {"Small": 8, "Medium": 9, "Large": 2, "Extra Large": 3}
```

```python
size_df["Size_Encoded_map"] = size_df["Size"].map(size_mapping)

size_df
```

**Removing Outlier using Zscore and IQR**


```python
q1 = loan_dataset["ApplicantIncome"].quantile(0.25)
q3 = loan_dataset["ApplicantIncome"].quantile(0.75)
iqr = q3 - q1
print(q1,q3,iqr)
```

```python
min_range=q1-1.5*iqr
max_range=q3+1.5*iqr
print(min_range,max_range)
```

```python
new_data =loan_dataset[loan_dataset["ApplicantIncome"] <= max_range]
new_data
```

```python
new_data.shape
```

```python
sns.boxplot(x="ApplicantIncome",data=new_data)
plt.show()
```
